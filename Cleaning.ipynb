# EDA dhe Vizualizime

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.month_name()
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
monthly_sales = monthly_sales.sort_values('Month')

sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

plt.plot(monthly_sales['Month'], monthly_sales['Sales'],
                  marker='o', markersize=8, linewidth=3, color='#2e7d32')


plt.title('Total Monthly Sales Performance', loc='left', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales (USD)', fontsize=12)
plt.xticks(range(1, 13)) 
sns.despine()

plt.tight_layout()
plt.gca().yaxis.set_major_formatter(ticker.ScalarFormatter(useOffset=False))
plt.gca().ticklabel_format(style='plain', axis='y')
plt.show()

```

```python
# 1. Prepare data
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

# 2. Visual setup
plt.figure(figsize=(10, 6))
sns.set_theme(style="white") 

# 3. Use a conditional color palette to highlight the max year
colors = ['#A5D6A7' if (x < max(yearly_sales['Sales'])) else '#2E7D32' for x in yearly_sales['Sales']]

# 4. Create the bar plot
bars = plt.bar(yearly_sales['Year'], yearly_sales['Sales'], color=colors, width=0.7)

# 5. Add direct labels (so you don't need a Y-axis)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + (max(yearly_sales['Sales'])*0.01), 
             f'${yval/1e6:.1f}M', ha='center', va='bottom', fontweight='bold', color='#424242')

# 6. Styling the "Analyst" way
plt.title('Annual Revenue Growth', loc='left', fontsize=18, fontweight='bold', pad=25)
plt.ylabel('Total Sales (Millions USD)', fontsize=10, color='gray')
plt.xlabel('')

sns.despine(left=True)
plt.yticks([]) 

plt.tight_layout()
plt.show()
```

```python
# 1. Prepare data
monthly_profit = df.groupby('Month')['Profit'].sum().reset_index()

# 3. Setup the Visual
plt.figure(figsize=(10, 5))
sns.set_theme(style="whitegrid") # Grid helps track the profit levels

# 4. Plot the line
# We use a distinct color (Teal/Blue) to differentiate Profit from Sales (Green)
plt.plot(monthly_profit['Month'], monthly_profit['Profit'], 
         marker='o', linewidth=3, color='#00796b')

# 5. Add "M" labels to the points (Simplified Loop)
for i in range(len(monthly_profit)):
    x = monthly_profit['Month'][i]
    y = monthly_profit['Profit'][i]
    
    # Label as "1.2M"
    plt.text(x, y + (y * 0.01), f'${y/1e6:.1f}M', ha='center', fontweight='bold')

# 6. Final Polish
plt.title('Monthly Profit Analysis', loc='left', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Profit (Millions USD)')
sns.despine(left=True)

plt.tight_layout()
plt.gca().yaxis.set_major_formatter(ticker.ScalarFormatter(useOffset=False))
plt.gca().ticklabel_format(style='plain', axis='y')
plt.show()

```

```python
# 1. Set the stage
plt.figure(figsize=(10, 6))
sns.set_theme(style="white")

# 2. Create the scatter plot
# alpha=0.5 makes the dots 50% transparent
# s=30 makes the dots a bit smaller and cleaner
plt.scatter(df['Sales'], df['Profit'], alpha=0.5, s=30, color='#1a237e')

# 3. Add a "Trend Line" (The Beginner way)
# This draws a simple average line to show the direction of the data
m, b = np.polyfit(df['Sales'], df['Profit'], 1) # Calculate the math for the line
plt.plot(df['Sales'], m*df['Sales'] + b, color='red', linewidth=2, label='Trend')

# 4. Clean up the labels
plt.title('Relationship: Sales vs. Profit', loc='left', fontsize=16, fontweight='bold')
plt.xlabel('Sales (USD)', fontsize=12)
plt.ylabel('Profit (USD)', fontsize=12)

# 5. Remove the 1e6 scientific notation (The easy way)
plt.ticklabel_format(style='plain', axis='both') 

sns.despine()
plt.tight_layout()
plt.show()
```

```python
# 1. Prepare data
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()

# 2. Setup the Visual
plt.figure(figsize=(10, 6))
sns.set_theme(style="white")

# 3. Create a Horizontal Bar Chart (kind='barh')
# We use 'City' for y and 'Sales' for x
bars = plt.barh(top_cities['City'], top_cities['Sales'], color='#1565c0')

# 4. Invert the axis so the #1 city is at the top
plt.gca().invert_yaxis()

# 5. Add the "M" labels (Slightly different for horizontal)
for bar in bars:
    width = bar.get_width() # This is the "Sales" value
    label_text = f"{width/1000000:.1f}M"
    
    # x = width (end of bar), y = bar location
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             label_text, va='center', ha='left', fontsize=10, fontweight='bold')

# 6. Final Polish
plt.title('Top 10 Cities by Total Sales', loc='left', fontsize=16, fontweight='bold')
plt.xlabel('Sales (Millions USD)')

# Hide the X-axis numbers since we have labels on the bars
plt.xticks([]) 
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

```python
# 1. Prepare data (Sorted descending by default with value_counts)
payment_count = df['Payment Mode'].value_counts().reset_index()
payment_count.columns = ['Mode', 'Count']

# 2. Setup
plt.figure(figsize=(10, 5))
sns.set_theme(style="white")

# 3. Create Horizontal Bar Plot
bars = sns.barplot(data=payment_count, x='Count', y='Mode', palette="Blues_r")

# 4. Add the labels directly on the bars
for i, bar in enumerate(payment_count['Count']):
    plt.text(bar, i, f' {bar:,}', va='center', fontweight='bold', color='#333333')

# 5. Final Styling
plt.title('Preferred Payment Methods', loc='left', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Number of Orders', color='gray')
plt.ylabel('') 

sns.despine(left=True, bottom=True)
plt.xticks([]) 

plt.tight_layout()
plt.show()
```

```python
# 1. Prepare data
payment_sales = df.groupby('Payment Mode')['Sales'].sum()

# 2. Identify the largest value to "explode"
explode = [0] * len(payment_sales)

# We find the position (index) of the maximum value and set it to 0.1
largest_index = payment_sales.argmax() 
explode[largest_index] = 0.1 

# 3. Create the Standard Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(payment_sales, 
        labels=payment_sales.index, 
        autopct='%1.1f%%',   
        explode=explode,      
        startangle=140)       

# 4. Title and Clean up
plt.title('Sales Distribution by Payment Mode')
plt.axis('equal')             

plt.show()
```

```python

# 1. Prepare and Sort data
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False).reset_index()

# 2. Setup
plt.figure(figsize=(10, 6))
sns.set_theme(style="white")

# 3. Create the bar plot
bars = sns.barplot(data=category_sales, x='Category', y='Sales', palette='viridis')

# 4. Add the "M" labels (Simple Loop)
for i, val in enumerate(category_sales['Sales']):
    label_text = f"${val/1000000:.1f}M"
    plt.text(i, val, label_text, ha='center', va='bottom', fontweight='bold')

# 5. Final Polish
plt.title('Sales Performance by Product Category', loc='left', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Total Sales (Millions USD)')
plt.xlabel('') 

# Clean up the Y-axis numbers to remove the 1e6
plt.yticks([]) 
sns.despine(left=True)

plt.tight_layout()
plt.show()
```

```python
# 1. Setup
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid") # Grid helps see the discount levels clearly

# 2. Plot the dots
# s=50 makes the dots a good size, alpha=0.5 handles the overlap
plt.scatter(df['Discount'], df['Profit'], alpha=0.5, s=50, color='#d32f2f')

# 3. Add a "Zero Profit" baseline
# This makes it obvious which discounts are causing losses
plt.axhline(0, color='black', linestyle='--', linewidth=1.5)

# 4. Clean up labels and title
plt.title('Impact of Discounts on Profitability', loc='left', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Discount (0.0 to 1.0)', fontsize=12)
plt.ylabel('Profit (USD)', fontsize=12)

# 5. Remove scientific notation
plt.ticklabel_format(style='plain', axis='y')

sns.despine()
plt.tight_layout()
plt.show()
```

# Preparing the Machene Learing DataSet

```python
ml_df = df.copy()
```

```python
ml_df["Year"] = ml_df["Order Date"].dt.year
ml_df["Month"] = ml_df["Order Date"].dt.month
ml_df["YearMonth"] = ml_df["Order Date"].dt.to_period('M').astype(str)
```

```python
ml_df = ml_df.drop(columns=[
    'Order ID',
    'Customer Name',
    'Product Name',
    'Order Date',
    'Month_Name',
    'YearMonth',
    'Profit',
    'Profit_log'
])
```

```python
ml_df.to_csv("ecommerce_ml.csv", index=False)
```
